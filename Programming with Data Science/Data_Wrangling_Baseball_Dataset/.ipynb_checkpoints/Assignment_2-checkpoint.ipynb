```python
## Q3. How many players have hit 40 or more HRs in one single season? (Number only)

df1 = df[['yearID', 'playerID', 'HR']].groupby(['yearID', 'playerID'], as_index= False).sum()
len(df1[df1['HR']>=40].drop_duplicates('playerID'))
```

```python
## Q4. How many players have hit 600 or more HRs for their career? 
##    (Dataframe with player first and last name and amount of HRs)

df2 = df.groupby(['playerID', 'nameFirst', 'nameLast']).agg({'HR': 'sum'})
df3 = df2[df2['HR']>=600].sort_values(by= 'HR').reset_index()
df3[['nameFirst', 'nameLast','HR']]
```

```python
## Q5. How many unique players have hit 40 2Bs or more, 10 3Bs or more, 200 Hits or more, and 30 HRs or more in one season? 
##    (Number Only)?

df4 = df.groupby(['playerID', 'yearID' ], as_index= False).agg({'HR': 'sum', '2B': 'sum', '3B' : 'sum', 'H' : 'sum'})
df5 = df4[(df4['HR']>=30) & (df4['2B']>=40) & (df4['3B']>=10) & (df4['H']>=200)].reset_index()
df5['playerID'].nunique()
```

```python
## Q6. How many players seasons have had 100 or more SBs? 
##    (Dataframe, first name, last name, yearID, number of SBs, ordered from descending SBs)

SeasonGroup = df.groupby(['yearID','playerID'], as_index = False).agg({'nameFirst' : 'first' ,'nameLast' : 'first', 'SB' : 'sum'})
SeasonList = SeasonGroup[SeasonGroup['SB'] >= 100].sort_values(by='SB', ascending= False).reset_index()
SeasonList[['nameFirst','nameLast','yearID','SB']]
```

```python
## Q7. How many players in the 1960s (yearID) have hit 200 or more HRs? 
##    (Dataframe, first name, last name, number of HRs, ordered descending by HR amt)

df7 = df[(df['yearID']>= 1960) & (df['yearID']< 1970)].groupby(['playerID', 'nameFirst', 'nameLast'], as_index= False).agg({'HR': 'sum'})
df7[df7['HR']>=200].sort_values(by= 'HR', ascending = False).reset_index()
#df8[['nameFirst', 'nameLast', 'HR']].sort_values(by= 'HR', ascending = False)
```

```python
## Q8. Who has hit the most HRs in history? (Dataframe, first name, last name, HRs)

df9 = df.groupby(['playerID', 'nameFirst', 'nameLast']).agg({'HR': 'sum'}).reset_index()
df9[df9['HR']==df9['HR'].max()][['nameFirst', 'nameLast', 'HR']]
```

```python
## Q9. Who had the most hits (H) in the 1970s? (Dataframe)

df10 = df[(df['yearID']>=1970) & (df['yearID']<=1979)].groupby(['playerID', 'nameFirst', 'nameLast']).agg({'H': 'sum'}).reset_index()
df10[df10['H']==df10['H'].max()]
```

```python
## Q10. Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  
##     (Dataframe, first name, last name, OBP, ordered descending by OBP) 
## OBP = (Hits + Walks(BB) + Hit by Pitch) / (At Bats + Walks(BB) + Hit by Pitch + Sacrifice Flies)
## Plate Appearance : PA = AB + BB + HBP + SH + SF + Times Reached on Defensive Interference(Not taken as not available in dataset)


df['OBP'] = (df['H']+df['BB']+df['HBP']).div (df['AB']+df['BB']+df['HBP']+df['SF'])
df['PA'] = (df['AB']+df['BB']+df['HBP']+df['SH']+df['SF'])
df.fillna(value=0, inplace = True)

df11 = df[(df['yearID']== 1977) & (df['PA']>= 500)].sort_values(by='OBP', ascending = False).head(5)
df11[['nameFirst', 'nameLast', 'OBP']]
```

```python
## Q11. Top 8 highest averages in 2013 with at least 300 PAs? (Dataframe, first name, last name, average, descending by average)
##      Batting Average = Hits/ AB

df['Bat_Avg'] = df['H']/df['AB']
df.fillna(value=0)

df12 = df.groupby(['yearID','playerID'], as_index= False).agg({'nameFirst': 'first', 'nameLast': 'first', 'PA': 'sum', 'Bat_Avg': 'sum'}).sort_values(by='PA', ascending=False)
df12 = df12[(df12['PA'] >=300) & (df12['yearID']== 2013)].sort_values(by='Bat_Avg', ascending= False).head(8)
df12[['nameFirst', 'nameLast', 'Bat_Avg']].reset_index()
```

```python
## Q12. Leaders in hits from 1940 up to and including 1949. (Dataframe, first name, last name, number of hits) 

df13= df[(df['yearID']>=1940) & (df['yearID']<=1949)].groupby(['playerID'], as_index= False).agg({'nameFirst': 'first', 'nameLast': 'first', 'H': 'sum'}).sort_values(by='H', ascending = False)
df13[['nameFirst', 'nameLast', 'H']].reset_index().head(5)
```

```python
## Q13.  Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number of hits)

df14 = df.groupby(['yearID','playerID'], as_index= False).agg({'nameFirst': 'first', 'nameLast': 'first', 'H': 'sum'}).sort_values(by='H', ascending= False)
df15 = df14.groupby(['yearID'], as_index= False).first().sort_values(by=['yearID','H'], ascending= [True,False])
df16 = df15.groupby(['playerID', 'nameFirst', 'nameLast'], as_index= False).agg({'H' : 'count'}).sort_values(by= 'H', ascending = False)
df16 = df16.rename(columns = {'H': 'Number_of_Times(Most_Hits)'})
df16.reset_index().head(2)

## Both 'Pete Rose' and 'Ichiro Suzuki' had most hits and most number of times (6 times each) 
```

```python
## Q14. Which players have played the most games for their careers?  
##      Top 5 first name, last name, descending by games played presented as a dataframe

df17 = df.groupby(['playerID', 'nameFirst', 'nameLast'], as_index= False).sum().sort_values(by='G', ascending= False)
df17[['nameFirst', 'nameLast', 'G']].reset_index().head(5)
```

```python
## Q15. How many players have had more 3000 or more hits for their careers while also hitting 500 or more HRs?  
##      Just a number

df18 = df[['H', 'HR','playerID']]
df19 = df18.groupby(['playerID'], as_index = False).sum().sort_values(by= 'HR', ascending = False)
len(df19[(df19['H']>= 3000) & (df19['HR']>=500)])
```

```python
## Q16. How many HRs were hit during the entire 1988 season?  Just a number

df20 = df[df['yearID']== 1988]
df20['HR'].sum()
```

```python
## Q17. Filter out and show the top 3 Batting average seasons by Wade Boggs during his career 
##      in seasons in which he had at least 500 or more ABs.  
##      Dataframe, first name, last name, average, descending by average

df21 = df[(df['nameFirst']== 'Wade') & (df['nameLast']== 'Boggs') & (df['AB']>= 500)].sort_values(by='Bat_Avg', ascending = False)
df21[['nameFirst', 'nameLast', 'Bat_Avg']].reset_index().head(3)
```

```python
## Q18.Please filter out the top 10 OBPs for the 1995 season with 400 or more PAs, sorted by OBP descending.  
##     Dataframe with first name, last name, OBP

df22 = df[(df['yearID']== 1995) & (df['PA']>= 400)].sort_values(by='OBP', ascending = False)
df22[['nameFirst', 'nameLast', 'OBP']].reset_index().head(10)
```

```python
## Q19. Who had the most 3Bs (in total) in 1922, 1925, 1926, and 1928?  
##      Dataframe with first name, last name, number of 3Bs

df23 = df[(df['yearID']== 1922) | (df['yearID']== 1925) | (df['yearID']== 1926) | (df['yearID']== 1928)]
df24 = df23.groupby(['yearID','playerID', 'nameFirst', 'nameLast'], as_index= False).sum().sort_values(by='3B', ascending= False)
df25 = df24.groupby(['yearID'], as_index= False).first().sort_values(by='yearID', ascending= True)
df25[['nameFirst', 'nameLast','yearID', '3B']]
```

```python
## Q20. How many unique players have hit 30 or more HRs in season while also stealing (SB) 30 more or bases?  
##      A number is okay here

df26 = df.groupby(['yearID','playerID', 'nameFirst', 'nameLast'], as_index= False).sum()
df27 = df26[(df26['HR']>= 30) & (df26['SB']>= 30)]
df27['playerID'].nunique()
```

```python
## Q21. Who had the highest OBP is 1986 with 400 or more ABs? (Dataframe first name, last name, OBP)

df28 = df.groupby(['yearID','playerID', 'nameFirst', 'nameLast'], as_index= False).sum()
df29 = df28[(df28['AB']>= 400) & (df28['yearID']== 1986)].sort_values(by= 'OBP', ascending = False)
#df29
df29[['nameFirst', 'nameLast', 'OBP']].head(1)
```

```python
## Q22.  Same question but for 1997 and only in the NL (check league ID)? (Dataframe, first name, last name OBP)

df30 = df.groupby(['yearID','playerID', 'nameFirst', 'nameLast', 'lgID'], as_index= False).sum()
df31 = df30[(df30['lgID']== 'NL') & (df30['yearID']== 1997)].sort_values(by= 'OBP', ascending = False)
df31[['nameFirst', 'nameLast', 'OBP']].head(3)

## There are 3 players with OBP = 1.0 in this question
```

```python
## Q23. Who had more than the league average HRs (for players with 500 or more ABs) in 2012 
##     (filter out all players with less 500 ABs)? (Dataframe first name, last name, HR descending)

mean_HR = df['HR'].mean()
df32 = df[df['yearID']== 2012]
df33 = df32.groupby(['yearID','playerID', 'nameFirst', 'nameLast'], as_index= False).sum()
df34 = df33[(df33['HR'] > mean_HR) & (df33['AB']>= 500)].sort_values(by= ['HR', 'AB'], ascending = [False, False])
df34[['nameFirst', 'nameLast', 'HR', 'AB']].reset_index()
```

```python
## Q24. Who is the youngest player to hit 50 or more HRs in a single season? 
##      (Dataframe, first name, last name, HRs, season)

df['player_age'] = df['yearID']- df['birthYear'] 
df35 = df.groupby(['yearID','playerID', 'nameFirst', 'nameLast'], as_index= False).sum()
#df35
df36 = df35[(df35['HR']>= 50)].sort_values(by= 'player_age')
df36[['nameFirst', 'nameLast', 'HR', 'yearID', 'player_age']].head(1)
```

```python
## Q25. Who are the five youngest players to hit 300 or more HRs for their career? 
##     (Dataframe, first name, last name, season they eclipsed more than 300 HRs)

## I have taken Cumulative sum of HR to find when a player get their earlist 300 or more HR.

df37 = df.groupby(['yearID', 'playerID', 'nameFirst', 'nameLast'], as_index= False).agg({'HR' : 'sum', 'player_age': 'first'}).sort_values(by='yearID')

df37['cum_HR'] = df37.groupby(['playerID', 'nameFirst', 'nameLast'], as_index= False)['HR'].cumsum()
df38 = df37[df37['cum_HR']>=300].sort_values(by='player_age')
df39 = df38.groupby(['playerID', 'nameFirst', 'nameLast'], as_index= False).agg({'player_age': 'first', 'cum_HR': 'first', 'yearID': 'first'}).sort_values(by=['player_age', 'cum_HR'], ascending=[True, False])
df39.reset_index().head(5)
```

```python
## Q26. Graph total HRs per season using bar graph

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df40 = df.groupby(['yearID'], as_index= False).sum()
df41 = df40[['yearID', 'HR']] 
plt.figure(figsize=(18,8), dpi=80, facecolor='w', edgecolor='r')
plt.bar(df41['yearID'], df41['HR'], align = 'center')
plt.xlabel('Years from 1871 to 2016')
plt.ylabel('Sum of HR per Season')
plt.show()
```

```python
## Q27. Using a line graph please graph the average HRs per AB (think about this) per season

df42 = df.groupby(['yearID'], as_index= False).sum()
df43 = df42[['yearID', 'HR', 'AB']]
df43['HR/AB'] = df43['HR']/df43['AB']
df43

plt.figure(figsize=(18,8), dpi=80, facecolor='w', edgecolor='r')
plt.bar(df43['yearID'], df43['HR/AB'], align = 'center', color='g')
plt.xlabel('Years from 1871 to 2016')
plt.ylabel('Average HRs per AB')
plt.show()
```
